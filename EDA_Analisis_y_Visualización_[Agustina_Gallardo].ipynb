```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("data.csv", encoding="latin-1")
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

df.head()
```

## Hipótesis 1
A mayor cantidad de productos vendidos, menor es el precio unitario.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Quantity"], y=df["UnitPrice"])

plt.title("Relación entre cantidad y precio unitario")
plt.xlabel("Cantidad")
plt.ylabel("Precio unitario")

plt.show()
```

### Análisis

El gráfico muestra la relación entre la cantidad de productos vendidos y el precio unitario.
No se observa una tendencia clara entre ambas variables, ya que los datos se encuentran dispersos.

### Conclusión

La hipótesis no puede ser confirmada con los datos disponibles.

## Hipótesis 2
La mayoría de las ventas se concentran en un solo país.

```python
plt.figure(figsize=(10,5))
sns.countplot(y=df["Country"], order=df["Country"].value_counts().index)

plt.title("Cantidad de transacciones por país")
plt.xlabel("Cantidad")
plt.ylabel("País")

plt.show()
```

### Análisis

El gráfico muestra la cantidad de transacciones por país.
Se observa que un país concentra la mayor cantidad de ventas, mientras que el resto tiene menor participación.

### Conclusión

La hipótesis se confirma.

## Hipótesis no comprobable

Las promociones o descuentos influyen en el aumento de ventas.

### Explicación

No es posible validar esta hipótesis ya que el dataset no contiene información sobre promociones, descuentos o campañas de marketing.
